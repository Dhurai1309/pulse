```python
#Aggregated insurance data

Path_1 = "data/aggregated/insurance/country/india/state/"
aggregated_Insurance_stateList = os.listdir(Path_1)

column_1 = {"State": [], "Year": [], "Quarter": [], "Insurance Type": [], "Insurance Count": [], "Insurance Amount": []}

output_folder = "Data_Extract"

for state in aggregated_Insurance_stateList:
    Current_State = Path_1 + state + "/"
    aggregated_Year_List = os.listdir(Current_State)
    
    for year in aggregated_Year_List:
        Current_Year = Current_State + year + "/"
        aggregated_jsonFile_List = os.listdir(Current_Year)
        
        for jsonFile in aggregated_jsonFile_List:
            Current_File = Current_Year + jsonFile
            data = open(Current_File, "r")

            A = json.load(data)

            for i in A["data"]["transactionData"]:
                name = i["name"]
                count = i["paymentInstruments"][0]["count"]
                amount = i["paymentInstruments"][0]["amount"]
                column_1["Insurance Type"].append(name)
                column_1["Insurance Count"].append(count)
                column_1["Insurance Amount"].append(amount)
                column_1["State"].append(state)
                column_1["Year"].append(year)
                column_1["Quarter"].append(jsonFile.split(".")[0])

aggregated_Insurance = pd.DataFrame(column_1)

aggregated_Insurance["State"] = aggregated_Insurance["State"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar")
aggregated_Insurance["State"] = aggregated_Insurance["State"].str.replace("-"," ")
aggregated_Insurance["State"] = aggregated_Insurance["State"].str.title()
aggregated_Insurance['State'] = aggregated_Insurance['State'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")

os.makedirs(output_folder, exist_ok=True)
output_file = os.path.join(output_folder, "Aggregated_insurance.csv")
aggregated_Insurance.to_csv(output_file, index=False)
```

```python
#Top insurance data

Path_7 = "data/top/insurance/country/india/state/"
Top_Insurance_stateList = os.listdir(Path_7)

column_7 = {"State": [], "Year": [], "Quarter": [], "Level": [], "Entity Name": [], "Insurance Count": [], "Insurance Amount": []}

output_folder = "Data_Extract"

for state in Top_Insurance_stateList:
    Current_State = Path_7 + state + "/"
    Map_Year_List = os.listdir(Current_State)

    for year in Map_Year_List:
        Current_Year = Current_State + year + "/"
        map_jsonFile_List = os.listdir(Current_Year)

        for jsonFile in map_jsonFile_List:
            Current_File = Current_Year + jsonFile
            data = open(Current_File, "r")

            G = json.load(data)

            if G['data']['states'] is not None:
                for states in G["data"]["states"]: 
                    name = states["entityName"]
                    count = states["metric"]["count"]
                    amount = states["metric"]["amount"]
                    column_7["Level"].append('State')
                    column_7["Entity Name"].append(name)
                    column_7["Insurance Count"].append(count)
                    column_7["Insurance Amount"].append(amount)
                    column_7["State"].append(state)
                    column_7["Year"].append(year)
                    column_7["Quarter"].append(jsonFile.split(".")[0])

            if G['data']['districts'] is not None:
                for districts in G["data"]["districts"]: 
                    name = districts["entityName"]
                    count = districts["metric"]["count"]
                    amount = districts["metric"]["amount"]
                    column_7["Level"].append('District')
                    column_7["Entity Name"].append(name)
                    column_7["Insurance Count"].append(count)
                    column_7["Insurance Amount"].append(amount)
                    column_7["State"].append(state)
                    column_7["Year"].append(year)
                    column_7["Quarter"].append(jsonFile.split(".")[0])
           
            if G['data']['pincodes'] is not None:
                for pincodes in G["data"]["pincodes"]:
                    name = pincodes["entityName"]
                    count = pincodes["metric"]["count"]
                    amount = pincodes["metric"]["amount"]
                    column_7["Level"].append('Pincode')
                    column_7["Entity Name"].append(name)
                    column_7["Insurance Count"].append(count)
                    column_7["Insurance Amount"].append(amount)
                    column_7["State"].append(state)
                    column_7["Year"].append(year)
                    column_7["Quarter"].append(jsonFile.split(".")[0])

top_Insurance = pd.DataFrame(column_7)

top_Insurance["State"] = top_Insurance["State"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar")
top_Insurance["State"] = top_Insurance["State"].str.replace("-"," ")
top_Insurance["State"] = top_Insurance["State"].str.title()
top_Insurance['State'] = top_Insurance['State'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")

os.makedirs(output_folder, exist_ok=True)
output_file = os.path.join(output_folder, "Top_insurance.csv")
top_Insurance.to_csv(output_file, index=False)
```
